```python
# Program to read the entire content from a text file and display it.

file = open("sample.txt", "r")

content = file.read()

print(content)

file.close()
```

```python
# Program to read the first n lines from a text file.

n = int(input("Enter number of lines: "))

file = open("sample.txt", "r")

for i in range(n):
    line = file.readline()

    if line == "":
        break

    print(line, end="")

file.close()
```

```python
# Program to accept input from the user and append it to a text file.

text = input("Enter text: ")

file = open("sample.txt", "a")

file.write(text)
file.write("\n")

file.close()

print("Data appended successfully.")
```

```python
# Program to read contents from a text file line by line and store each line into a list.

file = open("sample.txt", "r")

lines = []

for line in file:
    lines.append(line.strip())

file.close()

print(lines)
```

```python
# Program to find the longest word from the text file.

file = open("sample.txt", "r")

content = file.read()

file.close()

words = content.split()

longest = ""

for word in words:
    if len(word) > len(longest):
        longest = word

print("Longest word:", longest)
```

```python
# Program to count the frequency of a user-entered word in a text file.

word = input("Enter the word to search: ")

file = open("sample.txt", "r")

content = file.read()

file.close()

words = content.split()

count = 0

for w in words:
    if w == word:
        count += 1

print("Frequency:", count)
```
