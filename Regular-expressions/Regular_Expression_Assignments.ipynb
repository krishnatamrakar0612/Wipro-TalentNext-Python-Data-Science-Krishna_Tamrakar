```python
#Regex 1. Check for Octal Digits

import re

values=["789","123","004"]

for s in values:
    if re.fullmatch(r"[0-7]+",s):
        print(s,"-> Valid")
    else:
        print(s,"-> Invalid")
```

```python
#Regex 2. Extract User ID, Domain and Suffix

import re

emails="""zuck@facebook.com
sunder33@google.com
jeff42@amazon.com"""

for email in emails.split():
    m=re.match(r"(\w+)@(\w+)\.(\w+)",email)
    print(m.groups())
```

```python
#Regex 3. Split Irregular Sentence

import re

sentence="A, very    very; irregular_sentence"
words=re.split(r"[,;_\s]+",sentence)
print(" ".join(words))
```

```python
#Regex 4. Clean Tweet

import re

tweet="Good advice! RT @TheNextWeb: What I would do differently if I was learning to code today http://t.co/lbwej0pxOd cc: @garybernhardt #rstats"

tweet=re.sub(r"http\S+","",tweet)
tweet=re.sub(r"RT\s+","",tweet)
tweet=re.sub(r"@\w+","",tweet)
tweet=re.sub(r"#\w+","",tweet)
tweet=re.sub(r"cc:","",tweet)
tweet=re.sub(r"[^A-Za-z\s]","",tweet)
tweet=" ".join(tweet.split())
print(tweet)
```

```python
#Regex 5. Extract Text Between HTML Tags

import requests
import re

url="https://raw.githubusercontent.com/selva86/datasets/master/sample.html"
html=requests.get(url).text

text=re.findall(r">([^<>]+)<",html)
result=[i.strip() for i in text if i.strip()]
print(result)
```

```python
#Regex 6. Words Beginning and Ending with Same Character

import re

words=["civic","trust","widows","maximum","museums","aa","as"]

for word in words:
    if re.fullmatch(r"(.).*\1",word):
        print(word)
```
