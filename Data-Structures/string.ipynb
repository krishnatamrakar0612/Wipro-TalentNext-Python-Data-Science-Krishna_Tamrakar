```python
# String 1: Count upper and lower case letters in a string
s = "Wipro Technologies"
upper = sum(1 for ch in s if ch.isupper())
lower = sum(1 for ch in s if ch.islower())
print("Upper:", upper)
print("Lower:", lower)
```

```python

# String 2: Check whether a given string is a palindrome or not
s = "radar"
print(s == s[::-1])
```

```python
# String 3: Return n copies of first 2 chars where n is length of string
s = "Wipro"
result = s[:2] * len(s)
print(result)
```

```python
# String 4: Remove leading/trailing 'x' if present
s = "xHix"
if s.startswith('x'):
    s = s[1:]
if s.endswith('x'):
    s = s[:-1]
print(s)
```

```python

# String 5: Return n repetitions of the last n characters
s = "Wipro"
n = 3
print(s[-n:] * n)
```
