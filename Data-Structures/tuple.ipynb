```python
# Tuple 1: Print the 4th element from first and 4th element from last in a tuple
tuple_values = (10, 20, 30, 40, 50, 60, 70, 80)
print(tuple_values[3])
print(tuple_values[-4])
```

```python
# Tuple 2: Check whether an element exists in a tuple or not
tuple_values = (10, 20, 30, 40, 50)
element = 30
print(element in tuple_values)
```

```python

# Tuple 3: Convert a list into a tuple
sample_list = [10, 20, 30, 40, 50]
converted_tuple = tuple(sample_list)
print(converted_tuple)
```

```python

# Tuple 4: Find the index of an item in a tuple
tuple_values = (10, 20, 30, 40, 50)
item = 40
print(tuple_values.index(item))
```

```python
# Tuple 5: Replace last value of tuples in a list to 100
sample_list = [(10, 20, 40), (40, 50, 60), (70, 80, 90)]
updated_list = [values[:-1] + (100,) for values in sample_list]
print(updated_list)
```
