```python
# Set 1: Remove a given item from the set
sample_set = {10, 20, 30, 40, 50}
sample_set.discard(30)
print(sample_set)
```

```python

# Set 2: Create an intersection of sets
set_a = {10, 20, 30, 40}
set_b = {30, 40, 50, 60}
intersection = set_a.intersection(set_b)
print(intersection)
```

```python

# Set 3: Create a union of sets
set_a = {10, 20, 30}
set_b = {30, 40, 50}
union = set_a.union(set_b)
print(union)
```

```python

# Set 4: Find the maximum and minimum value in a set
sample_set = {5, 15, 25, 35, 45}
print(max(sample_set))
print(min(sample_set))
```
