```python
# List 1: Create a list of 5 integers, display items, and access individual elements by index
numbers = [10, 20, 30, 40, 50]
print(numbers)
print(numbers[0])
print(numbers[1])
print(numbers[2])
print(numbers[3])
print(numbers[4])
```

```python
# List 2: Append a new item to the end of the list
numbers = [10, 20, 30, 40, 50]
numbers.append(60)
print(numbers)
```

```python
# List 3: Reverse the order of the items in the list
numbers = [10, 20, 30, 40, 50]
numbers.reverse()
print(numbers)
```

```python
# List 4: Print the number of occurrences of a specified element in a list
numbers = [10, 20, 30, 20, 40, 20, 50]
count = numbers.count(20)
print(count)
```

```python
# List 5: Append the items of list1 to list2 in the front
list1 = [1, 2, 3]
list2 = [4, 5, 6]
list2 = list1 + list2
print(list2)
```

```python
# List 6: Insert a new item before the second element in an existing list
numbers = [10, 20, 30, 40, 50]
numbers.insert(1, 15)
print(numbers)
```

```python
# List 7: Remove the item from a specified index in a list
numbers = [10, 20, 30, 40, 50]
del numbers[2]
print(numbers)
```

```python
# List 8: Remove the first occurrence of a specified element from a list
numbers = [10, 20, 30, 20, 40, 50]
numbers.remove(20)
print(numbers)
```
