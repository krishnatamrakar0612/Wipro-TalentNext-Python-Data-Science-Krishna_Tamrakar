```python
# Dictionary 1: Add a key and value to a dictionary
sample_dict = {0: 10, 1: 20}
sample_dict[2] = 30
print(sample_dict)
```

```python
# Dictionary 2: Concatenate dictionaries to create a new one
dic1 = {1: 10, 2: 20}
dic2 = {3: 30, 4: 40}
dic3 = {5: 50, 6: 60}
new_dict = {}
new_dict.update(dic1)
new_dict.update(dic2)
new_dict.update(dic3)
print(new_dict)
```

```python

# Dictionary 3: Check if a given key already exists in a dictionary
sample_dict = {1: 10, 2: 20, 3: 30}
key = 2
if key in sample_dict:
    print("Key exists in the dictionary")
else:
    print("Key does not exist in the dictionary")
```

```python
# Dictionary 4: Iterate over a dictionary and print keys, values, and key-value pairs
sample_dict = {1: 10, 2: 20, 3: 30}

print("Keys:")
for key in sample_dict:
    print(key)

print("Values:")
for value in sample_dict.values():
    print(value)

print("Keys and Values:")
for key, value in sample_dict.items():
    print(key, value)
```

```python
# Dictionary 5: Prepare a dictionary with keys 1 to 15 and square values
squares = {number: number * number for number in range(1, 16)}
print(squares)
```

```python

# Dictionary 6: Sum all the values in a dictionary
sample_dict = {1: 10, 2: 20, 3: 30, 4: 40}
total = sum(sample_dict.values())
print(total)
```
