```python
# 1. Check whether a number is positive, negative, or zero
number = int(input('Enter a number: '))
if number > 0:
    print('Positive')
elif number < 0:
    print('Negative')
else:
    print('Zero')
```

```python
# 2. Check whether a number is odd or even
number = int(input('Enter a number: '))
if number % 2 == 0:
    print('Even')
else:
    print('Odd')
```

```python
# 3. Given two non-negative values, print true if they have the same last digit
first = int(input('Enter first number: '))
second = int(input('Enter second number: '))
if first % 10 == second % 10:
    print('true')
else:
    print('false')
```

```python
# 4. Print numbers from 1 to 10 in a single row with tab space
for value in range(1, 11):
    print(value, end='\t')
```

```python
# 5. Print even numbers between 23 and 57, one per line
for value in range(23, 58):
    if value % 2 == 0:
        print(value)
```

```python
# 6. Check whether a given number is prime
number = int(input('Enter a number: '))
if number < 2:
    print('Not Prime')
else:
    is_prime = True
    for divisor in range(2, int(number ** 0.5) + 1):
        if number % divisor == 0:
            is_prime = False
            break
    if is_prime:
        print('Prime')
    else:
        print('Not Prime')
```

```python
# 7. Print prime numbers between 10 and 99
for number in range(10, 100):
    is_prime = True
    if number < 2:
        is_prime = False
    else:
        for divisor in range(2, int(number ** 0.5) + 1):
            if number % divisor == 0:
                is_prime = False
                break
    if is_prime:
        print(number)
```

```python
# 8. Print the sum of all digits of a given number
number = int(input('Enter a number: '))
total = 0
while number > 0:
    total += number % 10
    number //= 10
print(total)
```

```python
# 9. Reverse a given number and print it
number = int(input('Enter a number: '))
reversed_number = 0
original_number = number
while number > 0:
    reversed_number = reversed_number * 10 + number % 10
    number //= 10
print(reversed_number)
```

```python
# 10. Check whether a number is a palindrome
number = int(input('Enter a number: '))
original_number = number
reversed_number = 0
while number > 0:
    reversed_number = reversed_number * 10 + number % 10
    number //= 10
if original_number == reversed_number:
    print(f'{original_number} is a palindrome.')
else:
    print(f'{original_number} is not a palindrome.')
```
