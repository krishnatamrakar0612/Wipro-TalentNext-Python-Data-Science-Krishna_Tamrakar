```python
# Command Line Arguments 3: Accept 10 numbers through command line arguments and calculate the sum of prime numbers among them
import sys

def is_prime(number):
    if number < 2:
        return False
    for divisor in range(2, int(number ** 0.5) + 1):
        if number % divisor == 0:
            return False
    return True

if len(sys.argv) != 11:
    print("Usage: python script.py <10 numbers>")
else:
    numbers = [int(value) for value in sys.argv[1:]]
    prime_sum = sum(number for number in numbers if is_prime(number))
    print(prime_sum)
```
